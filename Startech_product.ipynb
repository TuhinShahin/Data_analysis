```python
from bs4 import BeautifulSoup
import requests

def scrape_laptops(start_url, max_pages=50):
    laptops = []

    for page in range(1, max_pages + 1):
        url = f"{start_url}&page={page}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        # Find all laptop items
        laptop_items = soup.find_all("div", class_="p-item")

        # Extract data for each laptop item
        for item in laptop_items:
            title_element = item.find("h4", class_="p-item-name")
            description = title_element.text.strip()
            product_url = title_element.find("a")["href"]
            image_url = item.find("div", class_="p-item-img").find("img")["src"]
            specifications = item.find("div", class_="short-description").text.strip()
            price = item.find("div", class_="p-item-price").text.strip()

            # Store data in a dictionary
            laptop_data = {
                "Description": description,
                "Product_url": product_url,
                "Image_url": image_url,
                "Specifications": specifications,
                "Price": price
            }

            laptops.append(laptop_data)

    return laptops

# URL of the search page
start_url = "https://www.startech.com.bd/product/search?&search=laptop"

# Scrape data for all laptops (50 pages)
all_laptops = scrape_laptops(start_url, max_pages=50)

# Print scraped data
for laptop in all_laptops:
    print(laptop)
```

```python
def preprocess_laptops(laptops):
    for laptop in laptops:
        # Extract additional features from specifications
        specs = laptop['Specifications']

        # Initialize extracted features
        ram = None
        storage_capacity = None
        processor = None
        screen_size = None

        # Extract RAM size
        try:
            ram_str = specs.split('RAM:')[1].split(',')[0].strip()
            ram = int(ram_str.split()[0]) if ram_str.split()[0].isdigit() else None
        except IndexError:
            pass

        # Extract storage capacity
        try:
            storage_str = specs.split('Storage:')[1].split(',')[0].strip()
            storage_capacity = int(storage_str.split()[0]) if storage_str.split()[0].isdigit() else None
        except IndexError:
            pass

        # Extract processor type
        try:
            processor = specs.split('Processor:')[1].split(',')[0].strip()
        except IndexError:
            pass

        # Extract screen size
        try:
            screen_size_str = specs.split('Display:')[1].split(',')[0].strip()
            screen_size = int(screen_size_str.split('"')[0]) if screen_size_str.split('"')[0].isdigit() else None
        except IndexError:
            pass

        # Add extracted features to the laptop dictionary
        laptop['RAM'] = ram
        laptop['storage_capacity'] = storage_capacity
        laptop['processor'] = processor
        laptop['screen_size'] = screen_size

# Preprocess the scraped laptop data
preprocess_laptops(all_laptops)
```

Creating the database with SQlite and storing the data accordingly

```python
import sqlite3

# Connect to SQLite database (creates the database if it doesn't exist)
conn = sqlite3.connect('laptops.db')

# Create a cursor object to execute SQL queries
cur = conn.cursor()

# Define a SQL query to create a table
create_table_query = '''
CREATE TABLE IF NOT EXISTS laptops (
    Description TEXT,
    Product_url TEXT,
    Image_url TEXT,
    Specifications TEXT,
    Price REAL,
    RAM REAL,
    storage_capacity REAL,
    processor TEXT,
    screen_size REAL
)
'''

# Execute the create table query
cur.execute(create_table_query)

# Insert data into the table
for laptop in all_laptops:
    cur.execute('''
        INSERT INTO laptops (Description, Product_url, Image_url, Specifications, Price, RAM, storage_capacity, processor, screen_size)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (laptop['Description'], laptop['Product_url'], laptop['Image_url'], laptop['Specifications'], laptop['Price'], laptop['RAM'], laptop['storage_capacity'], laptop['processor'], laptop['screen_size']))

# Commit changes and close connection
conn.commit()
conn.close()
```
